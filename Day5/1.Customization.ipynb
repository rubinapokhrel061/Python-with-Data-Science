```python
# Title
import matplotlib.pyplot as plt 
year =[1950,1960,1970,1990]
pop=[2.538,4.57,8.62,10.85]
plt.plot(year,pop)
plt.xlabel("year")
plt.ylabel("Population")   
plt.title("World Population Projections")
plt.show() 
```

```python
# Ticks
import matplotlib.pyplot as plt 
year =[1950,1960,1970,1990]
pop=[2.538,4.57,8.62,10.85]
plt.plot(year,pop)
plt.xlabel("year")
plt.ylabel("Population")   
plt.title("World Population Projections")
plt.yticks([0,2,4,6,8,10])
plt.show() 
```

```python

import matplotlib.pyplot as plt 
year =[1950,1960,1970,1990]
pop=[2.538,4.57,8.62,10.85]
plt.plot(year,pop)
plt.xlabel("year")
plt.ylabel("Population")   
plt.title("World Population Projections")
plt.yticks([0,2,4,6,8,10],[0,'2B','4B','6B','8B','10B'])
plt.show() 
```

```python
year =[1950,1960,1970,1990]
pop=[2.538,4.57,8.62,10.85]
# add more data 
year=[1920,1930,1940]+year
pop=[1.3,2.3,4.5]+pop
plt.plot(year,pop)
plt.xlabel("year")
plt.ylabel("Population")   
plt.title("World Population Projections")
plt.yticks([0,2,4,6,8,10],[0,'2B','4B','6B','8B','10B'])
plt.show() 
```
