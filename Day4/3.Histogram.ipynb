```python
import matplotlib.pyplot as plt
values=[2.3,4.5,4.5,3.2,5.2,3.5,3.8,3.9,3.5]
plt.hist(values,bins=3)
plt.show()
# A histogram represents the distribution of numerical data.
# The x-axis shows the range of data divided into bins,
# while the y-axis indicates the frequency (count) of data points within each bin.
# By specifying bins=3, the data will be divided into three intervals.
# The heights of the bars will reflect how many values fall within each interval.
```
